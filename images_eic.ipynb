```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# 定义协方差矩阵
cov_matrix = np.array([[2.31671277e-02, 2.07558533e-02],
                       [2.07558533e-02, 2.51325912e+01]])

# 定义均值
mean = np.array([2.5, 30])

# 根据协方差矩阵估算合理的绘图范围
# 生成网格点
x, y = np.mgrid[1:4.1:.1, 5:56:1]
pos = np.dstack((x, y))

# 创建二元高斯分布
rv = multivariate_normal(mean, cov_matrix)

# 计算每个网格点的概率密度值
z = rv.pdf(pos)

# 绘制实心的密度图
plt.figure(figsize=(8, 6))
plt.contourf(x, y, z, levels=100, cmap='viridis')
cbar = plt.colorbar()
cbar.set_label(label='Density', fontsize=18)  # 设置 colorbar 的 label 和字体大小
cbar.ax.tick_params(labelsize=18)
plt.xlabel('LAI', fontsize = 18)
plt.ylabel('Cab ($ug/cm^2$)', fontsize = 18)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)
# plt.title('Bivariate Gaussian Density Plot')
plt.grid(False)
plt.savefig('images\\post_dens_bands2.png', dpi = 300, bbox_inches='tight')
```
